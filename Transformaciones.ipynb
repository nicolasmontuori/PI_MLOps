```python
# Genero el campo "id" para cada uno de los dataframes (amazon_df, disney_df, hulu_df y netflix_df)
amazon_df ["id"] = "a" + amazon_df ["show_id"]
disney_df ["id"] = "d" + disney_df ["show_id"]
hulu_df ["id"] = "h" + hulu_df ["show_id"]
netflix_df ["id"] = "n" + netflix_df ["show_id"]
```

```python
# Creo la columna plataforma
amazon_df["plataforma"] = "amazon"
disney_df["plataforma"] = "disney"
hulu_df["plataforma"] = "hulu"
netflix_df["plataforma"] = "netflix"
```

```python
# Reemplazo valores nan del campo rating por g
amazon_df['rating'].fillna("g", inplace=True)
disney_df['rating'].fillna("g", inplace=True)
hulu_df['rating'].fillna("g", inplace=True)
netflix_df['rating'].fillna("g", inplace=True)
```

```python
# Cambio a formato datetime
amazon_df ["date_added"] = pd.to_datetime (amazon_df["date_added"], errors= "coerce")
disney_df ["date_added"] = pd.to_datetime (disney_df ["date_added"], errors= "coerce")
hulu_df ["date_added"] = pd.to_datetime (hulu_df ["date_added"], errors= "coerce")
netflix_df ["date_added"] = pd.to_datetime (netflix_df ["date_added"], errors= "coerce")
```

```python
# Cambio el texto a minusculas 
amazon_df = amazon_df.applymap(lambda s: s.lower() if type(s) == str else s)
disney_df = disney_df.applymap(lambda s: s.lower() if type(s) == str else s)
hulu_df = hulu_df.applymap(lambda s: s.lower() if type(s) == str else s)
netflix_df = netflix_df.applymap(lambda s: s.lower() if type(s) == str else s)
```

```python
# Convierto el campo "duration" en "duration_int" y "duration_type"
amazon_df [["duration_int", "duration_type"]] = amazon_df ["duration"].str.extract("(\d+)\s*(\w+)")
disney_df [["duration_int", "duration_type"]] = disney_df ["duration"].str.extract("(\d+)\s*(\w+)")
hulu_df [["duration_int", "duration_type"]] = hulu_df ["duration"].str.extract("(\d+)\s*(\w+)")
netflix_df [["duration_int", "duration_type"]] = netflix_df ["duration"].str.extract("(\d+)\s*(\w+)")
```

```python
# Remplazo la palabra "seasons" por "season"
amazon_df ["duration_type"] = amazon_df ["duration_type"].replace("seasons", "season")
disney_df ["duration_type"] = disney_df ["duration_type"].replace("seasons", "season")
hulu_df ["duration_type"] = hulu_df ["duration_type"].replace("seasons", "season")
netflix_df ["duration_type"] = netflix_df ["duration_type"].replace("seasons", "season")
```

```python
# Cambio el tipo de dato de la columna "duration_int" a integer
amazon_df["duration_int"] = amazon_df["duration_int"].astype("Int64")
disney_df["duration_int"] = disney_df["duration_int"].astype("Int64")
hulu_df ["duration_int"] = hulu_df["duration_int"].astype("Int64")
netflix_df["duration_int"] = netflix_df ["duration_int"].astype("Int64")
```

```python
# Unifico los cuatro datasets
df_unificado = pd.concat([amazon_df, disney_df, hulu_df, netflix_df], ignore_index=True)
```

```python
# Verifico que las transformaciones se hayan realizado correctamente
df_unificado.head ()
```

```python
# Elimino las columnas showid y duration
df_unificado.drop("duration", axis=1, inplace=True)
df_unificado.drop("show_id", axis=1, inplace=True)
```

```python
# Unifico los datasets de la carpeta rating
df_rating_unificado = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8])
```

```python
# Cambio el nombre de la columna rating por score
df_rating_unificado.rename(columns={"rating":"score"}, inplace=True)
```

```python
from datetime import date
```

```python
# Cambio el formato de la columna "timestamp" a AAAA,mm,dd
df_rating_unificado.timestamp = df_rating_unificado ["timestamp"].apply (lambda x: date.fromtimestamp (x))
```

```python
# Verifico que los cambios se hayan realizado correctamente
df_rating_unificado.head ()
```

```python
# Unifico los datasets "df_unificado" y "df_rating_unificado" a partir de las columnas id y movieId.
# A partir de este vamos a entrenar posteriormente nuestro modelo de ML
df_general = df_unificado.merge (df_rating_unificado, left_on = "id", right_on = "movieId",how = "outer")
```

```python
df_general.head ()
```

```python
# Elimino la columna movieId porque es la misma que Id
df_general.drop("movieId", axis=1, inplace=True)
```

```python
# Creo un dataset con el score promedio para cada película (movieId).
df_score_promedio = df_rating_unificado.groupby(["movieId"]).mean("score")
```

```python
df_score_promedio.head ()
```

```python
# Elimino la columna userId que no me hace falta.
df_score_promedio.drop (["userId"], inplace = True, axis = 1)
```

```python
#Redondeo el score a 2 decimales
df_score_promedio["score"] = round(df_score_promedio["score"],2)
df_score_promedio.head()
```

```python
# Unifico los datasets "df_unificado" y "df_score_promedio".
# Este va a ser el dataset que suba a la API para las consultas.
df_plataformas_score = df_unificado.merge (df_score_promedio, left_on = "id", right_on = "movieId", how= "left")
```
